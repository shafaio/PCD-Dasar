```python
# melakukan import library
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
# melakukan import real.jpg dalam variabel img dengan library matplotlib
img = plt.imread('img/copy.jpg')
imgcv = cv.imread('img/real.jpg')
```

```python
print(img.shape) # menampilkan sebuah dimensi dari img dengan nilai 1280, 1023, 3
x, y, z = img.shape # menginisialisasi kan x untuk tinggi, y untuk lebar, dan z untuk RGB 
```

```python
plt.imshow(img) #menampilkan citra dari img
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY) 
# menginisialisasi variabel gray yang didalamnya melakukan konversi RGB ke dalam GRAY
plt.imshow(gray, cmap='gray')
# menampilkan hasil gambar dari konversi gambar yang berada pada variabel gray
```

```python
print(gray.max()) # menampilkan nilai maksimal dari gambar yang sudah di konversikan ke dalam grayscale 
print(gray.min()) # menampilkan nilai minimal dari gambar yang sudah di konversikan ke dalam grayscale
print(img) # menampilkan matriks citra dari gambar RGB
```

```python
sumGray = np.zeros(256).astype(int) 
# menginisialisasi variabel sumGray yang memanggil library numpy dengan function zeros yang bertipe data integer
```

```python
for i in range(x): # perulangan baris dalam varibel i dengan batasan nilai dari variabel x
    for j in range(y): # perulangan kolol dalam varibel j dengan batasan nilai dari variabel y
        val = gray[i, j] # menginisialisasi variabel val yang akan berisi dengan nilai pixel sesuai dengan perulangan
        sumGray [val] +=1 # memberi nilai pada setiap elemen pixel yang melalui perulangan
```

```python
grayValue = [i for i in range (256)]
```

```python
plt.bar(grayValue, sumGray)
# menampilkan grafik bar histogram dengan bantuan memanggil library plt dan sumbu x nilai dari data grayValue sedangkan sumbu y data sumGray 
```

```python
plt.bar(grayValue, sumGray/(x*y))
```

```python
ni = sumGray
print(ni)
# plt.imshow(sumGray)
```

```python
hi = ni / x*y # menginisialisai variabel hi yang berisi nilai dari hasil perhitungan derajat keabuan citra
print(hi) # menampilkan matriks dari variabel hi
```
