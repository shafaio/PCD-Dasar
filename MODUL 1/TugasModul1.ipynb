# TUGAS CONVERET GRAY TANPA LIBRARY

```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
image = plt.imread('img/real.jpg')
plt.imshow(image, cmap='gray')
tinggi, lebar ,RGB = image.shape
```

```python
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY) # convert gray dengan library
plt.imshow(gray, cmap='gray')
```

```python
gray # array gray
```

```python
result = np.zeros([tinggi, lebar]).astype(int) # membuat canvas baru
temp = 0
for i in range (tinggi):
    for j in range (lebar):
        for k in range (RGB):
            
            temp += image[i][j][k] # RGB tiap pixel di jumlahkan
        
        result [i, j] = temp/3 # rata-rata RGB di simpan pada canvas result
        temp = 0
```

```python
plt.figure(facecolor='white')
plt.subplot(121)
plt.imshow(gray, cmap='gray')
plt.title("With Library")
plt.subplot(122)
plt.imshow(result, cmap='gray')
plt.title("Without Library")
```
