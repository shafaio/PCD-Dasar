# MEDIAN FILTERING

```python
def funcMed(matrix): # inisialisasi fungsi funcMed dengan parameter matrix
    rows, col = matrix.shape 
    value = rows*col
    array = np.sort(matrix, axis=None)

    if (value%2==0):
        a = array[value//2]
        b = array[value//2-1]
        c =  (a+b)/2

    else:
        c = array[value//2]

    return c
```

```python
def median(image, Rows, Col):
    result = np.zeros(image.shape)
    rows, col =  int((Rows-1)/2), int((Col-1)/2)

    pad = np.pad(image, ((rows, Rows-(rows-1)), (col, Col-(col-1))), "constant")

    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            result [i, j] = funcMed(pad[i:i+Rows, j:j+Col])

    return result
```

```python
medFiltering3x3 = median(image, 3, 3)
medFiltering5x5 = median(image, 5, 5)
medFiltering7x7 = median(image, 7, 7)
```

```python
plt.figure(facecolor='white', figsize=[15, 7])
plt.subplot(131)
plt.title('MEDIAN FILTERING 3x3')
plt.imshow(medFiltering3x3, 'gray')
plt.subplot(132)
plt.title('MEDIAN FILTERING 5x5')
plt.imshow(medFiltering5x5, 'gray')
plt.subplot(133)
plt.title('MEDIAN FILTERING 7x7')
plt.imshow(medFiltering7x7, 'gray')
```

# MEAN FILTERING

```python
def funcMean(matrix):
    value = np.sum(matrix)
    rows, col = matrix.shape

    return (value/(rows*col))
```

```python
def mean(image, Rows, Col):
    result = np.zeros(image.shape)
    rows, col = int((Rows-1)/2), int((Col-1)/2)
    pad = np.pad(image, ((rows, Rows-(rows-1)), (col, Col-(col-1))), "constant")

    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            result [i, j] = funcMean(pad[i:i+Rows, j:j+Col])

    return result
    
```

```python
meanFiltering3x3 = mean(image, 3, 3)
meanFiltering5x5 = mean(image, 5, 5)
meanFiltering7x7 = mean(image, 7, 7)
```

```python
plt.figure(facecolor='white', figsize=[15, 7])
plt.subplot(131)
plt.title('MEAN FILTERING 3x3')
plt.imshow(meanFiltering3x3, 'gray')
plt.subplot(132)
plt.title('MEAN FILTERING 5x5')
plt.imshow(meanFiltering5x5, 'gray')
plt.subplot(133)
plt.title('MEAN FILTERING 7x7')
plt.imshow(meanFiltering7x7, 'gray')
```

# MODUS FILTERING

```python
def funcModus(matrix):
    value = np.sort(matrix, axis=None)
    index = np.zeros(np.max(value)+1, dtype=int)

    for i in range(value.shape[0]):
        index[value[i]] += 1

    maximum = np.max(index)
    for j in range(index.shape[0]):
        if(index[j] == maximum):
            return j
```

```python
def modus(image, Rows, Col):
    result = np.zeros(image.shape)
    rows, col = int((Rows-1)/2), int((Col-1)/2)
    pad = np.pad(image, ((rows, Rows-(rows-1)), (col, Col-(col-1))), "constant")

    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            result [i, j] = funcModus(pad[i:i+Rows, j:j+Col])

    return result
    
```

```python
modusFiltering3x3 = modus(image, 3, 3)
modusFiltering5x5 = modus(image, 5, 5)
modusFiltering7x7 = modus(image, 7, 7)
```

```python
plt.figure(facecolor='white', figsize=[15, 7])
plt.subplot(131)
plt.title('MODUS FILTERING 3x3')
plt.imshow(modusFiltering3x3, 'gray')
plt.subplot(132)
plt.title('MODUS FILTERING 5x5')
plt.imshow(modusFiltering5x5, 'gray')
plt.subplot(133)
plt.title('MODUS FILTERING 7x7')
plt.imshow(modusFiltering7x7, 'gray')
```
